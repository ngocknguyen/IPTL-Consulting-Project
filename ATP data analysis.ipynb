## Data Exploratory Analysis on Players based on Countries

```python
#plot to show countries with the most active, avid players from 2019 - 2020
plt.figure(figsize=(10,5), dpi=200)
ax = sns.countplot(x='flag_code', data = df,order= pd.value_counts(df['flag_code']).iloc[:20].index, palette='pastel')
plt.title('Top 20 countries with most male players from 2019-2021', weight='bold', fontsize=13)
plt.xlabel('Countries with Flag Codes',weight='bold', fontsize=11)
plt.ylabel('Count by Players',weight='bold', fontsize=11)

for i in ax.patches:
    ax.text (i.get_x() + i.get_width()/2,i.get_height()+ 20.,i.get_height(),
    horizontalalignment='center', fontsize = 8)

plt.savefig('atp_country.png')
```

## Are there any upset in Grand Slams for Men players?

### we consider upset if Loser Rank is less than 5 in ranking compared to Winner Rank

```python
upset = atp[['Tournament','Series', 'Round', 'Winner','Loser','WRank', 'LRank']]
```

```python
upset = upset[(upset.Series == 'Grand Slam')]
```

```python
len(upset)
```

```python
upset['upset'] = np.where(upset['WRank'] - upset['LRank'] >=5,1,0)
```

```python
upset.head()
```

```python
plt.figure(figsize=(12,6), dpi=200)
ax7= sns.countplot(x='upset',hue='Tournament', data = upset)
plt.title('Upset in Grand Slams from 2019-2021', weight='bold', fontsize=15)
plt.xlabel('Upset', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

for i in ax7.patches:
    ax7.text (i.get_x() + i.get_width()/2,i.get_height()+ 1.,i.get_height(),
    horizontalalignment='center', fontsize = 12)
    
```
